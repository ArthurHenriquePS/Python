```python
# Ex. 10 A partir da var word = 'engenharia de software'. Se a indice da word == 'e' or 'o', continue 

word = 'engenharia de software'
i = 0
lis = []

while i < len(word):
    l = word[i]
    if l == 'e' or l == 'o':
        i += 1
        continue
    if l == " ":
        i += 1
        continue
    lis.append(l)
    i += 1
print(lis)
```

```python
# Ex. 14 Crie um programa que solicite nome, curso e idade de 3 pessoas, armazene cada pessoa como uma
# lista dentro de uma lista_completa

lc = []
lp = []
i = 0

while i < 3:
    n = str(input(f"Digite o {i + 1}° nome: "))
    c = str(input(f"Digite o curso do {n}: "))
    id = int(input(f"Digite a idade do {n}"))
    lp = [n, c, id]
    lc.append(lp)
    i += 1
print(lc)
```
