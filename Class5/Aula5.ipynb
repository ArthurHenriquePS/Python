```python
# Ex. 5 Soma de valores digitados pelo usuário. Crie um algoritmo
# para solicitar 5 valores não inteiros, insira em uma lista. Após
# isso, crie uma estrutura para somar todos elementos desta lista

# Estrutura de entrada
s = []
r = 1
# Loop
for n in range(5):
    # Estrutura de Saída
    n = float(input(f"Digite um valor na {r}° posição: "))
    # Adicionando na lista
    s.append(n)
    # Adicionando rodada
    r += 1
# Calculando a média total da lista
m = sum(s) / len(s)
# Estrutura de saída
print(f"A média total é {m}")
```

```python
# Ex. 8 Vamos fazer um algoritmo para cadastro com média. Solicite 3
# informações: nome, nota1, nota2 três vezes. Após isso, calcule a média 
# de cada aluno e retorne uma lista contendo nome, nota1, nota2, media de cada aluno

# Estrutura: lista = [[nome, nota1, nota2, media],
#                     [nome, nota1, nota2, media]]

# Estrutura de entrada
l = []
lu = []
r = 1

# Loop
for i in range(3):
    # Estrutura de entrada
    n = str(input(f"Digite o nome da {r} ° posição: "))
    n1 = float(input("Digite a primeira nota: "))
    n2 = float(input("Digite a segunda nota: "))
    #  Calculo da média
    m = (n1 + n2) / 2
    # Adicionando na lista de cada aluno
    lu = [n, n1, n2, round(m, 2)]
    # Adicionando na lista geral
    l.append(lu)
    # Adicionando mais uma rodada
    r += 1

# Estrutura de Saída
print(l)
```

```python
# Ex. 9 Crie um algoritmo para simular um sistema de compras, onde deverá ser solicitado
# nome 100x. Após isso, crie uma estrutura para quebrar a solicitação de nomes. 
# Após isso, crie uma estrutura para solicitar preço, com isso, deixe o respectivo nome 
# e preço juntos. Finalmente, mostre o valor total dos produtos.

l = []
vt = 0


for i in range(100):
    n = str(input("Digite o nome: ").lower())
    
    if n == "exit":
        break

    p = float(input("Digite o valor do preço: "))
    vt += p

    ln = [n, p]
    l.append(ln)

print(l)
print("Valor total: ", vt)
        
```

```python
# Ex. 10 Crie um algoritmo para solicitar nome e nota para 4 alunos. Insira em uma estrutura com lista dentro de lista
# (nome, nota). Após isso, crie uma estrutura para mostrar os alunos aprovado com nota maior ou igual a 6, mostrando
# o respectivo nome e nota. Crie uma estrutura para mostrar os alunos de nota menor que 6 e mostre o respectivo nome
# e nota

l = []
r = 1

for i in range(4):
    n = str(input(f"Digite o nome do {r}° aluno: "))
    nt = float(input("Digite a nota do aluno: "))
    la = [n, nt]
    l.append(la)
    r += 1
print(l)
for a in l:
    if a[1] >= 6:
        print(f"Aluno {a[0]}: aprovado!")
    if a[1] < 6: 
        print(f"Aluno {a[0]}: reprovado!")
```

```python
# Ex. 11 - Crie um algoritmo para simular sistema com menu. o Usuario pode cadastrar,
# listar ou sair da estrutura

# Se opção 1, solicite nome, idade, cidade, e deixe estas informações em uma estrutura de lista dentro de lista
# Se opção 2, print o respectivo nome, idade e cidade da pessoa
# Se opção 3, crie uma estrutura para sair do código

l = []

while True:
    op = str(input("Digite a sua opção 1- cadastro, 2- mostrar, 3- sair").lower())
    if op == "cadastro" or op == "1":
        lu = []
        n = str(input("Digite o nome: "))
        i = int(input("Digite a idade: "))
        c = str(input("Digite a sua cidade: "))
        lu.append(n)
        lu.append(i)
        lu.append(c)
        l.append(lu)
    if op == "mostrar" or op == "2":
        print(l)
    if op == "sair" or op == "3":
        print("Até mais")
        break
```
