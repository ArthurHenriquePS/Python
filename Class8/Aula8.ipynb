```python
import pandas as pd

dados ={
    "produto": ["Notebook", "Teclado", "Mouse", "Monitor", "Webcam"],
    "vendas": [1200, 300, 100, 500, 60],
    "lucro": [ 800, 100, 30, 300, 20],
    "ano": ["2023", "2023", "2021", "2021", "2020"]
}
print(dados)

# Coloca os dados em uma tabela (facilitando a visualização do conteúdo)
data_frame = pd.DataFrame(dados)
data_frame
```

```python
# Cálculo média
import pandas as pd
# Calcula a média de acordo com os items da coluna "lucro"
media = data_frame["lucro"].mean()
# Calcula a mediana de acordo com os items da coluna "lucro"
mediana = data_frame["lucro"].median()
# Calcula o valor máximo de acordo com os items da coluna "lucro"
maximo = data_frame["lucro"].max()
# Calcula o valor mínimo de acordo com os items da coluna "lucro"
minimo = data_frame["lucro"].min()
print(f"Médio: {media}, Mediana: {mediana}, Máximo: {maximo}, Mínimo: {minimo}")
```

```python
# Try/Except - Utilizamos para que o except realiza uma ação, caso o try não seja falso ou não verdadeira
try:
    media = data_frame["lucro"].mean()
    print(f"A média de lucro é: {media}")
# Erro possível da chave do Dicionário
except KeyError:
    print("Esta coluna não existe! ")
```

```python
# Ex. 1 Crie um estrutura try/except para converter temperatura celsius em fahreinht 
# fahreinht = (celsius * 9/5) + 32
while True:
    try: 
        tc = float(input("Digite a temperatura em celsius: "))
        tf = (tc * 9/5) + 32
        print(f"Temperatura em celsius: {tc}, Temperatura em fahreinht: {tf}")
        break
    except ValueError:
        print("Valor não numérico")
```

```python
# Ex. 2 Crie uma estrutura com try/except a qual calcule e traga um erro para calcular a area
# retangulo e circulo com estrutura de input para o usuário. Além disso, o usuário deve escolher a figura geométrica

while True:
    try:
        o = str(input("Digite a forma geométrica 1- Retângulo 2- Circulo 3- Sair")).lower()
        if o == "retângulo":
            b = float(input("Digite o valor da base: "))
            a = float(input("Digite o valor da altura"))
            ar = b * a
            print(f"Base: {b}, Altura: {a}, Área: {round(ar, 2)}")
            break
        elif o == "circulo":
            r = float(input("Digite o valor do raio: "))
            ac = 3.14 * (r ** 2)
            print(f"Raio: {r}, Área: {round(ac, 2)}")
            break
        elif o == "sair":
            print("Saindo!")
            break
        else: 
            print("Por favor, selecione a opção correta")
    except ValueError: 
        print("Valor incorreto!")
```

```python
import pandas as pd
from matplotlib import pyplot as plt

dados = pd.read_excel("dados_falhas - missing values.xlsx")

# Plota um gráfico 
x = dados["UDI"]
y1 = dados["Temperatura Ar [K]"]
y2 = dados["Temperatura Processo [K]"]

plt.plot(x, y1, x, y2)
plt.xlabel("Amostras") # Títlo eixo X
plt.ylabel("Temperatura") # Título eixo y
plt.legend(["Ar", "Processo"]) # Cores da linhas
plt.title("Temperatura: Ar vs Processo") # Título Gráfico
plt.show()
```
