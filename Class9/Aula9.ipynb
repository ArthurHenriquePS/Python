```python
# Ex. 1

comando = 'start'

if comando == 'start':
    print("Iniciando")

elif comando == "stop":
    print("Parando")

else: 
    print("Entrada inválida!")
```

```python
# Ex. 1 - Match Case

match comando:
    case 'start':
        print("Iniciando!")
    case 'stop':
        print("Parando")
    case _: # Caso nenhum dos valores acima não aconteça
        print("Entrada Inválida")
```

```python
# Ex. 7
# Match/case e while 
# comando = sair - saindo da estrutura
# case = sair - encerrando
# case qualquer outra coisa - comando inválido

while True:
    comando = str(input("Digite o comando")).lower()
    match comando:
        case "sair":
            print("Saindo da estrutura!")
            break
        case _:
            print("Comando inválido")
```

```python
# Treinamento de uma rede neural MLP (classificação)
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("dados_falhas - missing values.xlsx")
plt.plot(df['UDI'], df["Temperatura Ar [K]"], df['UDI'], df["Temperatura Processo [K]"])
plt.show()
```

```python
# Definir as variáveis de entrada e saída
from sklearn.model_selection import train_test_split

df.columns
x = df[['Temperatura Ar [K]',
       'Temperatura Processo [K]', 
       'Velocidade Rotacao [rpm]', 
       'Torque [Nm]']]
y = df["Tipo da Falha"]

# Separar os dados para treino e teste

xtreino, xteste, ytreino, yteste = train_test_split(x, y, train_size= 0.7, random_state=42)
print(xtreino.shape)
print(ytreino.shape)
print(xteste.shape)
print(yteste.shape)

# Desenvolvimento do modelo MLP (multilayer perceptron)
from sklearn.neural_network import MLPClassifier
 
modelo_mlp = MLPClassifier(hidden_layer_sizes=[8 ,8], max_iter=100, activation='relu', random_state=42) # Para adicionar mais camadas [8, 8] 2 camadas com 8 neurônios cada
modelo_mlp.fir(xtreino, ytreino)
```
