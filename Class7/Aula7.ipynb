```python
# Ex. 6 Crie um algoritmo usando while true para simular um menu interativo. se opção for 1
# print olá, se opção igual a 2, print sair

while True:
    n = int(input("Digite um número: "))
    if n == 1:
        print("Olá")
    elif n == 2:
        print("Saindo!")
        break
```

```python
# Ex. 7 - Crie um algoritmo usando while true para ignorar valores negativos, Se o número for 0
# quebre a estrutura de repetição, caso ao contrário, envie cada númerp para uma lista

l = []
while True:
    n = int(input("Digite um número: "))
    if n == 0:
        print("Saindo!")
        break
    elif n < 1:
        continue
    l.append(n)
print(l)
```

```python
# Ex. 8 - Cire um algoritmo para solicitar nome, porém, mande apenas os nomes do que 3 caracteres para a lista de nome

ln = []
while True:
    n = str(input("Digite um nome"))
    if len(n) < 3:
        continue
    elif n == 'sair':
        print("Saindo!")
        break
    ln.append(n)
print(ln)
```

```python
# Ex. 9 - Crie um algoritmo com lista de notas válidas. Se -1, pare a estrutura, se nota < 0 ou nota > 10
# , continue. Ao final, calcule a média das notas

l = []
p = 1
while True: 
    n = float(input(f"Digite a {p}° nota: "))
    if n == -1:
        print("Saindo!")
        break
    elif n < 0 or n > 10.0:
        continue
    l.append(n)
    print(l)
m = sum(l) / len(l)
print(f"Média: {m}")

```

```python
# Ex. 10 - Vamos criar um sistema de menu simples.

# 1 - adicionar número
# 2 - remover último número
# 3 - mostrar lista
# 4 - sair

l = []

while True:
    o = int(input("""Digite uma opção  1 - adicionar número
                    2 - remover último número
                    3 - mostrar lista
                    4 - sair"""))
    if o == 1:
        n = int(input("Digite um número: "))
        l.append(n)
    elif o == 2:
        l.pop()
    elif o == 3:
        print(l)
    elif o == 4:
        print("Saindo!")
        break
```

```python
# Ex. 11 Filtrar palavras com letra proibida. Se aparecer palavra com x, não mostrar
# palavra na lista

l = []
while True:
    p = str(input("Digite uma palavra: ")).lower()
    if "x" in p:
        continue
    elif p == "sair":
        break
    l.append(p)
print(l)
```

```python
# Ex. 12 Criar um programa a qual ao selecionar opção 1, solicitar e enviar nomes para uma lista nomes, 
# opção 2, solicitar e enviar cpfs para uma lista_cpf e ao digitar opção 3, solicitar e enviar idade 
# # para lista_idade. Crie também uma estrutura para juntar todas as listas


ln = []
lcpf = []
li = []

while True:
    o = int(input("Digite uma opção \n1- nomes \n2- cpf \n3- Idade \n4- Mesclar"))
    if o == 1:
        n = str(input("Digite o nome: "))
        ln.append(n)
        print(f"Lista de nome {ln}")
    elif o == 2:
        cpf = str(input("Digite o CPF: "))
        lcpf.append(cpf)
        print(f"Lista de CPF: {lcpf}")
    elif o == 3: 
        i = int(input("Digite a sua idade: "))
        li.append(i)
        print(f"Lista de Idade: {li}")
    elif o == 4:
        lt = [ln, lcpf, li]
        print(f"Lista Total: {lt}")
        break
```
