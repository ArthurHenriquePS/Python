```python
# Variáveis de entrada

n1 = float(input("Digite a sua primeira nota: "))
n2 = float(input("Digite a sua segunda nota: "))
n3 = float(input("Digite a sua terceira nota: "))

# Calculo da média
nota_final = (n1 + n2 + n3) / 3

# Estrutura de saída

print("Primeira nota: ", n1)
print("Segunda nota: ", n2)
print("Terceira nota: ", n3)
print(f"Média: {nota_final:.01f}")
```

```python
# Variáveis de entrada

notas = []

qn = 0
nn = 1

p = int(input("Digite a quantidade de notas: "))

while qn < p:
    n = float(input(f"Digite a sua {nn}° nota:"))
    notas.append(n)
    qn += 1
    nn += 1
    


# Calculo da média
nota_final = sum(notas) / 3

# Estrutura de saída

print(notas)
print("Primeira nota: ", notas[0])
print("Segunda nota: ", notas[1])
print("Terceira nota: ", notas[2])
print("Média nota: ", nota_final)
```
