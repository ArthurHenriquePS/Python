```python
# Ex. 8 Crie um algoritmo para solicitar nome, sobrenome e cpf para o usuário
# se o nome e sobrenome estiverem corretos, mostre que será veridicado o cpf
# se o nome, sobrenome e cpf estiverem corretos, mostre acesso VIP permitido
# se nome ou sobrenome e cpf estiverem corretos, mostre que sera verificado. 
# Qualquer outra condição, acesso vip negado

#Estrutura de Entrada
n = str(input("Digite o seu nome: "))
sn = str(input("Digite o seu sobrenome"))
cpf = str(input("Digite o seu CPF: "))

nc = "jose"
snc = "moreira"
cpfc = "12356789-12"

if n == nc and sn == snc:
    print("Verificando CPF")
elif n == nc and sn == snc and cpf == cpfc:
    print("Acesso vip permitido")
elif n == nc or sn == snc and cpf == cpfc:
    print("Sera Verificado")
else: 
    print("Acesso Negado")
```

```python
# Ex. 10 Cálculo de desconte do INSS. Solicite o salário para o 
# Usuário e retorne o valor a ser descontado e o valor a ser recebido

# Faixa Salarial (R$)         Alíquota (%)          Parcela a Deduzir (R$)
# Até R$ 1621,00              7,5%                  -
# De 1621,01 até 2902,84      9,0%                  24,32
# De 2902,85 até 4354,27      12,0%                 111,40
# De 4354,28 até 8475,55      14,0%                 198,49

# Estrutura de Entrada
s = round(float(input("Digite o seu salário: ")), 2)

# Condição

if s <= 1621.00:
    # Calculo
    sfr = s * 0.075
    # Estrutura de Saída
    print(f"Valor a ser recebido é {round(sfr, 2)}")
elif s >= 1621.01 and s<= 2902.84:
    # Calculo
    sf = s * 0.09
    sfr = sf - 24.32
    # Estrutura de Saída
    print(f"Valor a ser recebido é {round(sfr, 2)}")
elif s >= 2902.85 and s <= 4354.27:
    # Calculo
    sf = s * 0.12
    sfr = sf - 111.40
    # Estrutura de Saída
    print(f"Valor a ser recebido é {round(sfr, 2)}")
elif s >= 4354.28 and s <= 8475.55:
    # Calculo
    sf = s * 0.14
    sfr = sf - 198.49
    # Estrutura de Saída
    print(f"Valor a ser recebido é {round(sfr, 2)}")
else: 
    # Estrutura de Saída
    print("Boa Sorte")
```

```python
# Ex.11 Classificação de crédito: solicite renda mensal e score de crédito
# Crie as seguintes regras: 

# Aprovado - renda => 3000 and score >= 700
# Analise Manual - renda => 2000 and score >= 600
# Caso contrário - negado 


# Estrutura de Entrada
rm = float(input("Digite a sua renda mensal: "))
s = float(input("Digite o seu score: "))

# Condição
if rm >= 700 and s >= 700:
    # Estrutura de Saída
    print("Aprovado!")
elif rm >= 2000 and s >= 600:
    # Estrutura de Saída
    print("Analise Manual")
else:
    # Estrutura de Saída
    print("Negado!")
```
