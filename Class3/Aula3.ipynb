```python
# Ex. 1 Sistema de login com nível de acesso. solicite usuário e senha.
# Se usuario é igual a admin, crie uma estrutura de condição aninhada
# para solicitar senha e se a mesma for "1234", mostre que o usuário
# terá acesso total. Caso o usuário insira senha incorreta, mostre
# senha incorreta. Caso usuário insira usuário incorreto, mostre
# usuário Incorreto.

# Estrutura de Entrada
u = str(input("Digite o nome do usuário: "))

# Condicional
if u == "admin":
    # Estrutura de Entrada
    s = str(input("Agora digite a senha do ADMIN: "))
    # Condicional
    if s == "1234":
        # Estrutura de Saída
        print("Acesso autorizado!")
    else:
        # Estrutura de Saída
        print("Senha incorreta!")
else:
    # Estrutura de Saída
    print("Usuário Incorreto!")
```

```python
# Ex. 2 Classificação de idade. solicite a idade, se a idade for maior ou
# igual a 18, crie uma estrutura de condição aninhada para verificar se idade
# é maior ou igual a 60, se for, mostre que é idoso, senão. mostre que é adulto.
# Se a idade for maior ou igual a 12, adolecente, caso ao contrário
# criança

# Estrutura de Entrada
i = int(input("Digite a sua idade: "))


# Condicional
if i >= 18:
    # Condicional
    if i >= 60:
        # Estrutura de Saída
        print("Idoso!")
    # Condicional
    else:
        # Estrutura de Saída
        print("Adulto!")
# Condicional
else:
    # Condicional
    if i >= 12:
        # Estrutura de Saída
        print("Adolecente")
    # Condicional
    else: 
        print("Criança!")
```

```python
# Ex. 3 Aprovação com distinção. solicite nota, se nota for maior ou igual a 6, crie condição
# aninhada para verificar se nota é maior ou igual a 9, se for aprovado com excelência. Se nota não for maior ou igual a 9
# Aprovado. Caso ao contrário, reprovado

# Estrutura de Entrada
n = int(input("Digite a sua nota: "))

# Condição
if n >= 6:
    # Condicional 
    if n >= 9:
        # Estrutura de Saída
        print("Aprovado com excelência!")
    # Condicional
    else:
        # Estrutura de Saída 
        print("Aprovado!")
# Condicional
else:
    # Estrutura de Saída 
    print("Reprovado")
```

```python
# Ex. 4 Verificação de número. Solicite número e verifique se é maior do que zero, se for, crie uma
# estrutura de condição aninhada para verificar se este número é par, se for, print positivo e par.
# Se não for, Positivo e ímpar. Se número for igual a zero, print zero, caso contrário negativo

# Estrutura de Entrada
n = int(input("Digite um número: "))

# Condicional
if n > 0:
    # Condicional
    if n % 2 == 0:
        # Estrutura de Saída
        print("Positivo e par")
    # Condicional
    elif n % 2 != 0:
        # Estrutura de Saída
        print("Positivo e ímpar")
# Condicional        
else:
    # Condicional
    if n == 0:
        # Estrutura de Saída
        print("Zero!")
    # Condicional
    else:
        # Estrutura de Saída  
        print("Negativo!")
```

```python
# Ex. 5 Sistema de desconto. solicite valor e se a pessoa é vip ou não. Se o valor maior ou igual 200
# crie estrutura de condição aninhada para verificar se a pessoa é vip, se for, ofereça 20% de
# desconto sobre o valor e mostre o valor a ser descontado e o valor final considerando o desconto, 
# Se não for vip, ofereça o desconto de 10%

# Estrutura de Entrada
v = int(input("Digite o valor: "))
vip = str(input("Você é vip? S/N")).upper()

# Condicional
if v >= 200:
    # Condicional
    if vip == "S":
        # Calculo de Desconto
        d = v * 0.80
        # Estrutura de Saída
        print(f"Desconto de 20%, Valor final de {d}")
# Condicional       
else:
    # Calculo de Desconto
    d = v * 0.90
    # Estrutura de Saída
    print(f"Desconto de 10%, Valor final de {d}")
```

```python
# Ex. 6 Crie um algoritmo para perguntar para o usuário qual o dia da semana, caso seja sábado,
# escreva dia de festa. Caso seja domingo, pergunte sobre a condição física do usuário, se estiver
# com dores de cabeça, print recuperando, então, precisa descansar. Caso contrário
# apenas descanse. Caso não seja sábado ou domingo, mostre trabalhando, trabalhando e trabalhando 

# Estrutura de Entrada
ds = str(input("Digite o dia da semana: ")).lower()

# Condicional
if ds == "sabado":
    # Estrutura de Saída
    print("Dia de festa!")
# Condicional
elif ds == "domingo":
    # Estrutura de Entrada
    cf = str(input("Qual a sua condição física? ")).lower()
    # Condicional
    if cf == "dor de cabeça":
        # Estrutura de Saída
        print("Recuperando!")
# Condicional
else:
    # Estrutura de Saída
    print("Trabalhando, Trabalhando e Trabalhando")
```
